```python
import numpy as np
import pandas as pd

from sklearn.linear_model import LinearRegression
from importlib import reload


# Read data from file
motor1_data = pd.read_csv('EMAX Avan Quad-Blade 2 Inch Prop and BETA FPV 1105 6000KV Motor.csv')
motor2_data = pd.read_csv('EMAX Avan Quad-Blade 2 Inch Prop and EMAX RS1108 5200KV Motor.csv')


```

```python
# Motor 1 -  BETA FPV 1105 6000KV
```

```python
# Create vector X
xmesh = np.linspace(min(motor1_data['Motor Electrical Speed (RPM)']), max(motor1_data['Motor Electrical Speed (RPM)']), 100)

# Create vector for polyfit
X = motor1_data.loc[:, "Motor Electrical Speed (RPM)"].values.reshape(-1, 1)
Y = motor1_data.loc[:, "Thrust (gf)"].values.reshape(-1, 1)

# Use poly fit and return polynomial equation
fit = np.polyfit(motor1_data['Motor Electrical Speed (RPM)'], motor1_data['Thrust (gf)'], 2)
equation = np.poly1d(fit)
print ("{0:.4e}x^2 {1:+.4e}) {2:+.4e}".format(*fit))


fig1 = plt.figure()

ax1 = fig1.add_subplot(111)
ax1.set_title('Thrust vs Motor Speed and ', fontsize=10, fontweight='bold')
ax1.grid(True)
yy=(max(Y)-min(Y))*0.8
color = 'tab:blue'
ax1.text(X[0], yy, "f(x) = {0:.4e}x^2 {1:+.4e}x {2:+.4e}".format(*fit), style='italic')
ax1.set_xlabel('Motor Speed (RPM)')
ax1.set_ylabel('Thrust (gf)')
ax1.plot(X, Y, color=color,marker="o", linestyle='None', label='Thrust (gf)')
ax1.plot(xmesh, equation(xmesh), '-r', label='Curve fit')
ax1.legend(loc='best')

```

```python

Column = "Torque (Nm)"


# Create vector X
xmesh = np.linspace(min(motor1_data['Motor Electrical Speed (RPM)']), max(motor1_data['Motor Electrical Speed (RPM)']), 100)

# Create vector for polyfit
X = motor1_data.loc[:, "Motor Electrical Speed (RPM)"].values.reshape(-1, 1)
Y = motor1_data.loc[:, Column].values.reshape(-1, 1)

# Use poly fit and return polynomial equation
fit = np.polyfit(motor1_data['Motor Electrical Speed (RPM)'], motor1_data[Column], 2)
equation = np.poly1d(fit)
print ("{0:.4e}x^2 {1:+.4e}) {2:+.4e}".format(*fit))

fig2 = plt.figure()
ax2 = fig2.add_subplot(111)
ax2.set_title('Torque vs Motor Speed ', fontsize=10, fontweight='bold')


ax2.grid(True)

yy=(max(Y)-min(Y))*0.9
color = 'tab:blue'
ax2.text(X[0], yy, "f(x) = {0:.4e}x^2 {1:+.4e}x {2:+.4e}".format(*fit), style='italic')

ax2.set_xlabel('Motor Speed (RPM)')
ax2.set_ylabel(Column)
ax2.plot(X, Y, color=color,marker="o", linestyle='None', label = Column)
ax2.plot(xmesh, equation(xmesh), '-r', label='Curve fit')

ax2.legend(loc='best')


plt.show()
```
